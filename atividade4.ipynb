```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('acidentes_dados.csv')
df
```

```python
df.info()
```

```python
# Calculate mean, median, and mode for 'Age' column
mean_values = df[['Age']].mean()
median_values = df[['Age']].median()
mode_values = df[['Age']].mode().iloc[0]

print("Statistics for Age column:")
print(f"Mean: {mean_values['Age']}")
print(f"Median: {median_values['Age']}")
print(f"Mode: {mode_values['Age']}")
```

```python
# Calculate variance, standard deviation, and range for 'Age' column
age_variance = df['Age'].var()
age_std_dev = df['Age'].std()
age_range = df['Age'].max() - df['Age'].min()

print(f"Variance of Age: {age_variance}")
print(f"Standard Deviation of Age: {age_std_dev}")
print(f"Range of Age: {age_range}")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df, x='Year', palette='viridis')
plt.title('Number of Crashes per Year')
plt.xlabel('Year')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=90)
plt.tight_layout()

# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge')

plt.show()
```
