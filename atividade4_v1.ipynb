```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('acidentes_dados.csv')
df
```

```python
df.info()
```

```python
# Calculando parâmetros estatísticos para a coluna AGE (Idade)

mean_values = df[['Age']].mean()
median_values = df[['Age']].median()
mode_values = df[['Age']].mode().iloc[0]
age_variance = df['Age'].var()
age_std_dev = df['Age'].std()
age_range = df['Age'].max() - df['Age'].min()

print("Estatísticas para a coluna AGE (Idade):")
print()
print(f"Média........: {mean_values['Age']}")
print(f"Mediana......: {median_values['Age']}")
print(f"Moda.........: {mode_values['Age']}")
print(f"Variancia....: {age_variance}")
print(f"Desvio Padrão: {age_std_dev}")
print(f"Amplitude....: {age_range}")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Histograma de Idade X Frequência de Acidentes')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

Os jovens, próximo aos 19 anos são o maior número de envolvidos nos acidentes no período avaliado entre 1989 e 2021.

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df, x='Year', palette='viridis')
plt.title('Numero de Acidentes por Ano')
plt.xlabel('Ano')
plt.ylabel('Numero de Acidentes')
plt.xticks(rotation=90)
plt.tight_layout()

# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge')

plt.show()
```

A tendência é de queda no número de acidentes ao longo dos anos no período entre 1989 e 2021.
