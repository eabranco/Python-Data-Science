# Importando a biblioteca PANDAS

```python
import pandas as pd
```

# Lendo a base de dados IRIS

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
df
```

```python
df.info()
```

# Resposta 1

```python
average_measurements_by_species = df.groupby('species').mean()
display(average_measurements_by_species)
```

# Resposta 2

```python
standard_deviation_by_species = df.groupby('species').std()
display(standard_deviation_by_species)
```

# Resposta 3

```python
observation_counts_by_species = df['species'].value_counts()
display(observation_counts_by_species)
```

# Resposta 4

```python
min_max_measurements_by_species = df.groupby('species').agg(['min', 'max'])
display(min_max_measurements_by_species)
```

# Resposta 5

```python
overall_average_measurements = df.mean(numeric_only=True)
display("MÃ©dia Geral:")
display(overall_average_measurements)
display("MÃ©dia por EspÃ©cie:")
display(average_measurements_by_species)
```
